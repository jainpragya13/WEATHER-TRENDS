```python
global_average = globaltemp['avg_temp'].rolling(10).mean()
local_average = localtemp['avg_temp'].rolling(10).mean()
```

```python
plt.plot(globaltemp['year'],global_average);
plt.title('Global Average Temperature')
plt.xlabel('years')
plt.ylabel('Average')
```

```python
plt.plot(localtemp['year'],local_average)
plt.title('Local Average Temperature')
plt.xlabel('Years')
plt.ylabel('Average')
```

```python
plt.plot(globaltemp['year'],global_average,label =" Global Temperature");
plt.plot (localtemp['year'],local_average,label = "Local temperature" );
plt.title('Line plot for comparing moving average of tempreture for global and New Delhi weather data')
plt.xlabel('Years')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.09), shadow=True, ncol=2)
plt.ylabel('Moving Average')
```
